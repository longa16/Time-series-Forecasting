```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')

print(df.head())
print(df.info())
```

```python
#EDA
df['date'] = pd.to_datetime(df['date'])

daily_sales = df.groupby('date')['sales'].sum()

plt.figure(figsize=(14, 6))
plt.plot(daily_sales.index, daily_sales.values, color='#1f77b4', linewidth=1)
plt.title('Ventes Totales Quotidiennes', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Nombre total d\'articles vendus', fontsize=12)
plt.grid(True, alpha=0.3)
#plt.show()
```

```python
#Prophet for time series forecasting 
from prophet import Prophet
from sklearn.metrics import mean_absolute_error

df_prophet = pd.DataFrame({'ds': daily_sales.index, 'y': daily_sales.values})

#split data to train/test
cutoff_date = df_prophet['ds'].max() - pd.Timedelta(days=90)
train = df_prophet[df_prophet['ds'] < cutoff_date]
test = df_prophet[df_prophet['ds'] > cutoff_date]

print(f"Entranement sur {len(train)} jours")
print(f"Test sur {len(test)} jours ")
print(train.tail())
```

```python
#model implementation
model = Prophet(yearly_seasonality=True, daily_seasonality=False)
model.fit(train)

future = model.make_future_dataframe(periods=90)

forecast = model.predict(future)
fig1 = model.plot(forecast)
plt.title("Prdiction des ventes (Ligne bleue) vs Ralit (Points noirs)")
plt.show()

plt.figure(figsize=(10, 5))

plt.plot(test['ds'], test['y'], label='Ralit', color='black')

forecast_test = forecast[forecast['ds'] >= cutoff_date]
plt.plot(forecast_test['ds'], forecast_test['yhat'], label='Prdiction Prophet', color='blue')
plt.title('Comparaison Ralit vs Prdiction (3 derniers mois)')
plt.legend()
plt.show()
```

```python
#model precision evaluation
import numpy as np 
from sklearn.metrics import mean_absolute_error

y_true = test['y'].values
y_pred = forecast_test['yhat'].values

mae = mean_absolute_error(y_true, y_pred)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print(f"--- RSULTATS DU MODLE PROPHET ---")
print(f"Erreur Moyenne (MAE) : {mae:.2f} articles")
print(f"Moyenne des ventes relles : {y_true.mean():.2f} articles")
print(f"Pourcentage d'erreur (MAPE) : {mape:.2f}%")
```
